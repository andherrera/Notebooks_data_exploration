## Principal Component Analysis

So far we used only some data dimensions for visualization. We now apply Principal Component Analysis (PCA) to project the four original dimensions into a two dimensional space.Also we apply Standard Scaler to normalize the variable (x - mean(x) / standard_deviation(x) )

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#take in x the values(Sepal length, Sepal with, Petal length, Petal width) and the class of target values
x = iris.loc[:, numerical_variables].values
y = iris.loc[:,['Species']].values
```

```python
x
```

[texto del vnculo](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html). Principal component analysis is affected by attribute scale so we normalize all the attributes by eliminating the mean and scaling to unit variance.

```python
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
#Define how many components 2 dimensional space
pca = PCA(n_components=2)
new_data = pca.fit_transform(x)
pca_iris = pd.DataFrame(data = new_data, 
                        columns = ['principal component 1', 'principal component 2'])
```

```python
pca_iris.head()
```

```python
#Add the target values
pca_iris_complete = pca_iris
pca_iris_complete['Species'] = iris[['Species']]
pca_iris_complete.head()
```

```python
plt.figure(figsize = (8,8))
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('Principal Component Analysis (2 Components)', fontsize = 20)
targets = dataset.target_names
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = pca_iris_complete['Species'] == target
    plt.scatter(pca_iris_complete.loc[indicesToKeep, 'principal component 1']
               , pca_iris_complete.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
# ax.legend(targets)
plt.axis([-3,3,-3,3])
plt.grid()
```

After this we can see that one class y completely separable from the others. Now because we applied PCA we have lost also some information, then we can check how much info we lose in the next step

```python
print("Explained Variance")
print("   Component 1 %3.2f"%(pca.explained_variance_ratio_[0]))
print("   Component 2 %3.2f"%(pca.explained_variance_ratio_[1]))
print("   Total Explained Variance %3.2f"%sum(pca.explained_variance_ratio_))
```

96% of the variance is a good value!. Suppose we have 30-50% then I am not representing most of the data. Now we can see the linear transformation, and the coefficientes used in the transformation for all the four dimensions.

Component 1 = 0.52 * First_Column - 0.26 * Second_Column + ...  
Component 2 = 0.37 * First_Column + 0.92 * Second_Column + ... 

Component 1 is influenced by column 1, 3 and 4  
Component 2 is influenced by column 1, and 2


```python
print("Components")
for i,c in enumerate(pca.components_):
    print("Component %d\t%s"%(i,str(c)))
```

```python
data = np.dot(x,np.transpose(pca.components_))
data[:5,:]
```

# t-SNE

Fix random state to replicate the results

```python
from sklearn.manifold import TSNE

perplexity=80
tsne = TSNE(n_components=2, verbose=1, perplexity=perplexity, n_iter=300, random_state=2867976)
# tsne = TSNE(n_components=2, verbose=1, perplexity=10, n_iter=300)
tsne_result = tsne.fit_transform(x)
```

```python
iris_tsne = pd.DataFrame({'x':tsne_result[:,0], 'y':tsne_result[:,1], 'Species':iris['Species']})
```

The difference with PCA(deterministic) in this case with tse(stochastic) we dont have a linear relation between the values and the originals.The random seed affect the final graph 

```python
iris_tsne
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('x', fontsize = 15)
ax.set_ylabel('y', fontsize = 15)
ax.set_title('t-SNE (2 Components)', fontsize = 20)
targets = dataset.target_names
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = iris_tsne['Species'] == target
    ax.scatter(iris_tsne.loc[indicesToKeep,'x'], iris_tsne.loc[indicesToKeep,'y'], c=color, s=50)
ax.legend(targets)
plt.axis([-3,3,-3,3])
ax.grid()
```
